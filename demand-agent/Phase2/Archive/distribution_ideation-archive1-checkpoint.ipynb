```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data preperation

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

### Create data frame from input file

```python
df = pd.read_csv('building_data.csv').interpolate().fillna(-1)
df['Time'] = pd.to_datetime(df['Time'], infer_datetime_format=True)
```

```python
df
```

### Create melted data frame

```python
melted = pd.melt(df, id_vars=['Time', 'Day', 'Price( $ per kWh)', 'PV (W)', 'Temperature in LA in 2012 (Celsius)'], var_name='Office Name', value_name='Power consumption (kWh)')
```

```python
def filter_rows_by_values(df, col, values):
    return df[~df[col].isin(values)]

melted = filter_rows_by_values(melted, "Power consumption (kWh)", [-1])
```

```python
melted
```

### Add time variables

```python
melted.insert(2, 'Day of Week', melted['Day'], True)
melted.insert(1, 'Month', melted['Time'].map(lambda x: x.month), True)
melted["Day"] = melted['Time'].map(lambda x: x.day)
melted.insert(3, 'Hour', melted['Time'].map(lambda x: x.hour), True)
```

```python
melted
```

### Encode and Scale data

```python
# encode categorical data
encoder = LabelEncoder()
encoding_columns = ["Day of Week", "Office Name"]
for column in encoding_columns:
    melted[column] = encoder.fit_transform(melted[column]).astype('float32')
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
# these columns are NOT scaled
scalable_columns = melted.columns.difference(["Time", "Office Name"])
melted[scalable_columns] = scaler.fit_transform(melted[scalable_columns])
```

### Resulting prepared data

```python
prepared_data = melted.reset_index(inplace=False, drop=True)
prepared_data
```

```python
# Graph of prepared data (graph repeated later)
fig, ax = plt.subplots(figsize=(25,10))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
ax.scatter(prepared_data.loc[:,'Time'],prepared_data.loc[:,'Power consumption (kWh)'],c=prepared_data.loc[:,'Office Name'],label='sampled data')
ax.legend(loc='upper center',fontsize='large',shadow=True)
plt.show()
```

# Basic Learning Model

```python
import tensorflow_probability as tfp
import tensorflow as tf
import tensorflow_probability as tfp
import math
from sklearn.utils import shuffle
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Activation, Layer, Input
```

### Universal Settings

```python
training_data_percentage = 0.8
training_data_amount = math.floor(len(prepared_data)*training_data_percentage)
columns_to_ignore = prepared_data.columns.difference(["Time", "Office Name"])
```

```python
epochs = 50
batch_size = 50
learning_rate = 0.0003
adamOptimizer = optimizers.Adam(learning_rate=learning_rate)
```

## Basic data set

### Prepare training and test basic data sets

```python
labels = prepared_data.loc[:,'Time']
train_labels = labels[:training_data_amount]
test_labels = labels[training_data_amount:]
```

```python
training_data = prepared_data[columns_to_ignore].values[:training_data_amount,:]
test_data = prepared_data[columns_to_ignore].values[training_data_amount:,:]
train_x = training_data[:,:-1]
train_y = training_data[:,-1]
test_x = test_data[:,:-1]
test_y = test_data[:,-1]
```

### Conduct training on basic data set

```python
model = Sequential()
model.add(Dense(12,input_shape=(7,),activation="tanh"))
model.add(Dense(12,activation="tanh"))
model.add(Dense(1,activation="linear"))
model.compile(loss='mse',optimizer=adamOptimizer,metrics=['mse'])
history_cache = model.fit(train_x,
                          train_y,
                          verbose=1, # write =1 if you wish to see the progress for each epoch
                          epochs=epochs,
                          batch_size=batch_size)
y_pred = model.predict(test_x)
```

### Graph result for basic data set

```python
fig, ax = plt.subplots(figsize=(25,10))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
ax.scatter(test_labels,prepared_data.loc[training_data_amount:,'Power consumption (kWh)'],c=prepared_data.loc[training_data_amount:,'Office Name'],label='sampled data')
ax.scatter(test_labels,y_pred,c='r',label='predicted values')
ax.legend(loc='upper center',fontsize='large',shadow=True)
plt.show()
print('Final cost: {0:.4f}'.format(history_cache.history['mse'][-1]))
```

## Shuffled data set

### Prepare training and test shuffled data sets

```python
### shuffle data
shuffled_data = shuffle(prepared_data).reset_index(drop=True)
```

```python
shuffled_labels = shuffled_data.loc[:,'Time']
shuffled_train_labels = shuffled_labels[:training_data_amount]
shuffled_test_labels = shuffled_labels[training_data_amount:]
```

```python
shuffled_training_data = shuffled_data[columns_to_ignore].values[:training_data_amount,:]
shuffled_test_data = shuffled_data[columns_to_ignore].values[training_data_amount:,:]
shuffled_train_x = shuffled_training_data[:,:-1]
shuffled_train_y = shuffled_training_data[:,-1]
shuffled_test_x = shuffled_test_data[:,:-1]
shuffled_test_y = shuffled_test_data[:,-1]
```

### Conduct training on shuffled data set

```python
shuffled_model = Sequential()
shuffled_model.add(Dense(12,input_shape=(7,),activation="tanh"))
shuffled_model.add(Dense(12,activation="tanh"))
shuffled_model.add(Dense(1,activation="linear"))
shuffled_model.compile(loss='mse',optimizer=adamOptimizer,metrics=['mse'])
shuffled_history_cache = shuffled_model.fit(shuffled_train_x,
                          shuffled_train_y,
                          verbose=1, # write =1 if you wish to see the progress for each epoch
                          epochs=epochs,
                          batch_size=batch_size)
shuffled_y_pred = shuffled_model.predict(shuffled_test_x)
```

### Graph result for shuffled data set

```python
fig, ax = plt.subplots(figsize=(25,10))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
ax.scatter(shuffled_test_labels,shuffled_data.loc[training_data_amount:,'Power consumption (kWh)'],c=shuffled_data.loc[training_data_amount:,'Office Name'],label='sampled data')
ax.scatter(shuffled_test_labels,shuffled_y_pred,c='r',label='predicted values')
ax.legend(loc='upper center',fontsize='large',shadow=True)
plt.show()
print('Final cost: {0:.4f}'.format(shuffled_history_cache.history['mse'][-1]))
```

# Time based data set

### Prepare training and test time based data sets

```python
training_dataframe = prepared_data.loc[prepared_data.Month < 0.9]
test_dataframe = prepared_data.loc[prepared_data.Month >= 0.9]
```

```python
train_labels = training_dataframe.loc[:,'Time']
test_labels = test_dataframe.loc[:,'Time']
```

```python
training_data = training_dataframe[columns_to_ignore].values
test_data = test_dataframe[columns_to_ignore].values
train_x = training_data[:,:-1]
train_y = training_data[:,-1]
test_x = test_data[:,:-1]
test_y = test_data[:,-1]
```

### Conduct training on time based data set

```python
model = Sequential()
model.add(Dense(12,input_shape=(7,),activation="tanh"))
model.add(Dense(12,activation="tanh"))
model.add(Dense(1,activation="linear"))
model.compile(loss='mse',optimizer=adamOptimizer,metrics=['mse'])
history_cache = model.fit(train_x,
                          train_y,
                          verbose=1, # write =1 if you wish to see the progress for each epoch
                          epochs=epochs,
                          batch_size=batch_size)
y_pred = model.predict(test_x)
```

### Graph result for time based data set

```python
fig, ax = plt.subplots(figsize=(25,10))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
ax.scatter(test_labels,test_dataframe.loc[:,'Power consumption (kWh)'],c=test_dataframe.loc[:,'Office Name'],label='sampled data')
ax.scatter(test_labels,y_pred,c='r',label='predicted values')
ax.legend(loc='upper center',fontsize='large',shadow=True)
plt.show()
print('Final cost: {0:.4f}'.format(history_cache.history['mse'][-1]))
```

### Graph all-time result for time based data set

```python
all_labels = np.concatenate((train_labels,test_labels))
all_x = np.concatenate((train_x, test_x))
all_y_pred = model.predict(all_x)
```

```python
fig, ax = plt.subplots(figsize=(25,10))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
ax.scatter(prepared_data.loc[:,'Time'],prepared_data.loc[:,'Power consumption (kWh)'],c=prepared_data.loc[:,'Office Name'],label='sampled data')
ax.scatter(all_labels,all_y_pred,c='r',label='predicted values')
ax.legend(loc='upper center',fontsize='large',shadow=True)
plt.show()
```
